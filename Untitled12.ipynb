```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

spill_risk=pd.read_csv('/content/oil_spill_risk_dataset_10000_rows for project.csv')


X=spill_risk.drop('spill_risk', axis=1) # Feature dataset

print(X)

Y=spill_risk['spill_risk'] # Target data

X_train, X_test, Y_train, Y_test=train_test_split(X, Y, random_state=42, test_size=0.3) # Data splitting and testing

spill_Model=RandomForestClassifier(n_estimators=300, random_state=42) # Randomforest model

spill_Model.fit(X_train, Y_train) # Model training

spill_prediction=spill_Model.predict(X_test) # Model prediction
print(spill_prediction)

# Performanace evaluation

print('Confusion matrix is:' , confusion_matrix(Y_test, spill_prediction)) #Confusion_matrix for performance evaluation

print('Classificaton Report is:', classification_report(Y_test, spill_prediction))

print('Accuracy Score is:', accuracy_score(Y_test, spill_prediction))

# Probability + Risk

new_data = [X_test.iloc[0]]

risk = spill_Model.predict(new_data)
prob = spill_Model.predict_proba(new_data)

labels = ["Low", "Medium", "High"]

print("Predicted Risk:", labels[risk[0]])

# Confusion matrix annotated diagram

print("Probabilities:", prob)

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Confusion matrix values (replace with your actual results if different)
conf_matrix = np.array([[77, 90, 0],
                        [1, 2065, 34],
                        [0, 139, 594]])

labels = ["Low", "Medium", "High"]

# Convert to percentages per actual class
row_sums = conf_matrix.sum(axis=1, keepdims=True)
conf_matrix_percent = conf_matrix / row_sums * 100  # percentage

plt.figure(figsize=(6,4))
sns.heatmap(conf_matrix_percent, annot=True, fmt='.1f', cmap="Blues",
            xticklabels=labels, yticklabels=labels, cbar_kws={'label': 'Percentage (%)'})

plt.title("Annotated Confusion Matrix (Percentages) for Oil Spill Risk Prediction")
plt.ylabel("Actual Risk Level")
plt.xlabel("Predicted Risk Level")

# Add checkmarks for correct predictions, crosses for misclassifications
for i in range(conf_matrix.shape[0]):
    for j in range(conf_matrix.shape[1]):
        if conf_matrix[i,j] > 0:
            if i == j:
                plt.text(j+0.5, i+0.5, "", color='green', fontsize=16, ha='center', va='center')
            else:
                plt.text(j+0.5, i+0.5, "", color='red', fontsize=16, ha='center', va='center')

plt.tight_layout()
plt.show()

```
